```python
model_version = '/home/santi/BA/final_code/results/checkpoint-2500'
model = XLMRobertaForSequenceClassification.from_pretrained(model_version, output_attentions=True)
tokenizer = XLMRobertaTokenizer.from_pretrained('xlm-roberta-base')

text = " "
#inputs = tokenizer.encode_plus(text, return_tensors='pt', add_special_tokens=True)
sentence_a = "Can you set an alarm for 10"
sentence_b = "establecer una alarma para 10"
inputs = tokenizer.encode_plus(sentence_a, sentence_b, return_tensors='pt', add_special_tokens=True)

input_ids = inputs['input_ids']
attention = model(input_ids)[-1]
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list)
model_view(attention, tokens)
```
